```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 6 distributions:
# exponential
# geometric
# logistic
# power
# triangular
# uniform
# I picked distrubutions I know and like, but they're all kind of 'easy'
# Maybe I need to take some more math classes, even these ones are hard.

exp_var = np.random.exponential(1,100)
plt.hist(exp_var, color = 'aqua')
plt.axvline(exp_var.mean())
plt.axvline(exp_var.mean() + exp_var.std(), linestyle = 'dashed')
plt.axvline(exp_var.mean() - exp_var.std(), linestyle = 'dashed')

plt.show()

# The mean is close to one from this sample. The data is skewed right but most of the values
# are concentrated between 0 and 1. The standard deviation is also 1. This measure seems a little
# less useful as the left vertical deviation line is almost flush with the egde of the data
```

```python
geo_var = np.random.geometric(.5,100)
plt.hist(geo_var, color = 'aqua')
plt.axvline(geo_var.mean())
plt.axvline(geo_var.mean() + geo_var.std(), linestyle = 'dashed')
plt.axvline(geo_var.mean() - geo_var.std(), linestyle = 'dashed')
plt.show()

# The geometric distribution models the number of trials that must be run in order to achieve success.
# On average, two tries! I think the mean for this model tells us a great deal. The standard deviation not
# much, since it doesn't communicate the skew or the concentration of data at 1 (should be about half.)
# Also, the left standard deviation line is actually beyond any real data point.
```

```python
log_var = np.random.logistic(0,1,100)
plt.hist(log_var, color = 'aqua')
plt.axvline(log_var.mean())
plt.axvline(log_var.mean() + log_var.std(), linestyle = 'dashed')
plt.axvline(log_var.mean() - log_var.std(), linestyle = 'dashed')
plt.show()

# Logistic function resembles the normal distribution but has heavier tails. It is more helpful in describing
# data that has more extreme values. One application is that it is used by the World Chess Federation to
# describe player performance. The mean is where it should be and makes sense and is useful. The standard deviation
# also looks unexcitingly similar to how it looks in the normal distribution.

```

```python
power_var = np.random.power(2,1000000)
plt.hist(power_var, color = 'aqua',bins = 100)
plt.axvline(power_var.mean())
plt.axvline(power_var.mean() + power_var.std(), linestyle = 'dashed')
plt.axvline(power_var.mean() - power_var.std(), linestyle = 'dashed')
print(power_var.mean())
plt.show()

# I wanted to use extra sample size and bins for this one because it looks cool. Apparently for this function
# 0 <= x <= 1, and the first parameter ('2') actually means we are looking at a power of (2 - 1) = 1, hence
# the linearity of this function. I was able to figure this out by playing with the paremeters and increasing
# the bin size and then reading the notes at scipy.org to verify this. The mean appears to be 2/3...I'm
# vaguely remembering this and I think I need to do some integrals to verify. The standard deviation isn't very
# helpful for this distribution
```

```python
tri_var = np.random.triangular(1,2,5,1000000)
plt.hist(tri_var, color = 'aqua', bins = 100)
plt.axvline(tri_var.mean())
plt.axvline(tri_var.mean() + tri_var.std(), linestyle = 'dashed')
plt.axvline(tri_var.mean() - tri_var.std(), linestyle = 'dashed')
plt.show()

# This one's a little silly. I was thinking it had something to do with triangular numbers when I chose it
# (1,3,6,10,15,21...). Instead it's just a piecewise function of two lines and P = 0 outside of the bounds.
# This is actually used often in ill-defined problems where the underlying distribution is not known and in
# simulations. I'm intrigued by this distribution and its uses, but the mean and standard deviations
# are very unhelpful/unnecessary for describing this data.
```

```python
uni_var = np.random.uniform(1,2,100)
plt.hist(uni_var, color = 'aqua')
plt.axvline(uni_var.mean())
plt.axvline(uni_var.mean() + uni_var.std(), linestyle = 'dashed')
plt.axvline(uni_var.mean() - uni_var.std(), linestyle = 'dashed')
plt.show()

# Thought I'd end this off with a bang. Uniform distribution. The mean should be right between the bounds
# (first 2 parameters) and the standard deviation should be halfway between the mean and the respective
# bounds. They're not very useful because they are trivial.
```

```python
n=10000
normal_one = np.random.normal(5,.5,n)
normal_two = np.random.normal(10,1,n)
normal_sum = normal_one + normal_two

plt.hist(normal_sum, color = 'aqua')
plt.axvline(normal_sum.mean())
plt.axvline(normal_sum.mean() + normal_sum.std(), linestyle = 'dashed')
plt.axvline(normal_sum.mean() - normal_sum.std(), linestyle = 'dashed')
plt.show()

# mean = 15 -- this makes intuitive sense
# sd = 1 -- this not as much...
```
